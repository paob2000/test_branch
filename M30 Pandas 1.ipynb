```python
# añadir una nueva columna
df['duration_num'] = df['duration']
```

```python
df['duration_unit'] = ""
```

```python
df.head()
```

```python
#  se debe cambiar el tipo de dato de la columna
df['duration_unit'] = df['duration_unit'].astype(pd.StringDtype())
```

```python
# cambiar varias columnas al tiempo
stringcols = df.select_dtypes(include='object').columns
df[stringcols] = df[stringcols].fillna('').astype(pd.StringDtype)
df.dtypes
```

```python
df.sample(5)
```

```python
# definimos primero todas las unidades como min
df['duration_unit'] = "min"
```

```python
df.head(3)
```

```python
# llevamos el datos
df['duration_unit'] = df['duration_unit'].where(df.duration_num.str.endswith('min'), 'season')
df.sample(5)
```

```python
#usamos el replace para quitar el resto 
df['duration_num'] = df['duration_num'].replace({' min':'',' Seasons':'','Season':'','s':''}, regex=True)
df.sample(5)
```

```python
df.dtypes
```

```python
#muestra un subconjunto para validar la data
df[['show_id','type','duration','duration_num','duration_unit']].sample(10)
```

```python
#cambia columna a tipo numerico
df['duration_num'] = pd.to_numeric(df['duration_num'])
df.dtypes
```

```python
df.describe()
```

```python
#Hay mas regitrostotales que de duretion_num por tanto localizamos los nulos
df.loc[df['duration_num'].isnull()]
```

```python
#Edicion de valores manualmente

df.at[5541, 'duration_unit'] = 'min'
df.at[5794, 'duration_unit'] = 'min'
df.at[5813, 'duration_unit'] = 'min'

df.at[5541, 'duration_num'] = 74
df.at[5794, 'duration_num'] = 84
df.at[5813, 'duration_num'] = 66

df.at[5541, 'duration'] = '74 min'
df.at[5794, 'duration'] = '84 min'
df.at[5813, 'duration'] = '66 min'
```
