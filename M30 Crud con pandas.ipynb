# CRUD CON PANDAS

```python
#nuevo dataframe
import pandas as pd

clase = [
    ('Jorge', 22, 'MX'),
    ('Anita', 31, 'MX'),
    ('Raul', 16, 'BR'),
    ('Sinon', 41, 'BR'),
    ('Diego', 33, 'PA'),
    ('Jorge', 35, 'PA'),
    ('Almendra', 35, 'COL'),
    ('Mary', 35, 'COL'),
    ('Peter', 35, 'EC'),
    ('Andrés', 35, 'EC'),
]

df = pd.DataFrame(clase, columns=['Nombre','Edad','País'])
df
```

```python
#Ordenar por edad
df.sort_values(by='Edad', ascending=True)
```

```python
# Llega un alumno nuevo
df.loc[len(df)] = ['Paulo', 48, 'COL']

df
```

```python
#Eliminar un registro basado en el indice
df = df.drop(df.index[6])
df
```
